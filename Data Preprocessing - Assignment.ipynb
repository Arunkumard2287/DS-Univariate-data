```python
import pandas as pd
```

```python
dataset = pd.read_csv('kidney_disease.csv')
```

```python
dataset
```

```python
def quanqual(dataset):
    quan = []
    qual = []

    for ColumnName in dataset.columns:
        #print(ColumnName)

        if (dataset[ColumnName].dtype=='O'):
            #print("qual")
            qual.append(ColumnName)
        else:
            #print("quan")
            quan.append(ColumnName) 
    return quan,qual

quan, qual = quanqual(dataset)
print(quan)
print(qual)
```

```python
dataset.isnull().sum()
```

```python
# data preprocessing for quan columns
import numpy as np
from sklearn.impute import SimpleImputer

# SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
dataset[quan] = imp.fit_transform(dataset[quan])
dataset = pd.DataFrame(dataset)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
#data preprocessing for qual columns
# SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
dataset[qual] = imp.fit_transform(dataset[qual])
dataset = pd.DataFrame(dataset)
```
